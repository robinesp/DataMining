```python
# Ridge regression
from sklearn.linear_model import Ridge

ridge = Ridge()
scores = cross_val_score(ridge, X, y)
print("Cross-validation scores: {}".format(scores))
print("Average cross-validation score: {:.2f}".format(scores.mean()))

scores_scaled = cross_val_score(ridge, X_scaled, y)
print("\nCross-validation scores (scaled): {}".format(scores_scaled))
print("Average cross-validation score (scaled): {:.2f}".format(scores_scaled.mean()))
```

For Ridge and  Lasso, plot a curve showing the effect of the training and test set performance ($R^2$) while increasing the degree of regularization for different penalties. For ElasticNet, plot a heatmap $alpha \times l1\_ratio \rightarrow R^2$ using test set performance.

```python
# Default shuffled split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, shuffle=True, random_state=0)
print("Size of training set: {}\nSize of test set: {}\n".format(X_train.shape[0], X_test.shape[0]))

range_01 = [0, 0.001, 0.01, 0.1, 0.25, 0.5, 0.75, 0.9, 0.99, 0.999, 1]
```

```python
# Linear regression with polynomial features
from sklearn.preprocessing import PolynomialFeatures
best_score = 0
lr_scores = []

for degree in [2,3,5,7]:
    poly = PolynomialFeatures(degree=degree, include_bias=False)
    X_poly = poly.fit_transform(X_scaled)
    
    X_poly_train, X_poly_test, y_train, y_test = train_test_split(X_poly, y, shuffle=True, random_state=0)

    reg = LinearRegression().fit(X_poly_train, y_train)
    score = reg.score(X_poly_test, y_test)
    lr_scores.append(score)
    if score > best_score:
        best_score = score
        best_parameters = {'degree': degree}

print("Best score: {:.6f}".format(best_score))
print("Best parameters: {}".format(best_parameters))
```

```python
# Ridge Regression (L2)
best_score = 0
ridge_train_scores = []
ridge_test_scores = []

for alpha in range_01:
    ridge = Ridge(alpha=alpha).fit(X_train, y_train)
    score = ridge.score(X_test, y_test)
    ridge_test_scores.append(score)
    ridge_train_scores.append(ridge.score(X_train, y_train))
    if score > best_score:
        best_score = score
        best_parameters = {'alpha': alpha}

fig1=plt.figure(figsize=(14,3))
fig1.show()
ax1=fig1.add_subplot(121)
ax1.plot(range_01,ridge_test_scores,c='b',ls='-',label='Test')
plt.legend(loc=2)
plt.title("Ridge Regression test regularization")
plt.xlabel("Alpha")
plt.ylabel("R^2 score")
ax2=fig1.add_subplot(122)
ax2.plot(range_01,ridge_train_scores,c='r',ls='-',label='Training')
plt.legend(loc=2)
plt.title("Ridge Regression training regularization")
plt.xlabel("Alpha")
plt.ylabel("R^2 score")
plt.draw()

fig2=plt.figure(figsize=(14,3))
fig2.show()
ax=fig2.add_subplot(111)
ax.plot(range_01,ridge_test_scores,c='b',ls='-',label='Test')
ax.plot(range_01,ridge_train_scores,c='r',ls='-',label='Training')
plt.legend(loc=2)
plt.title("Ridge Regression regularization")
plt.xlabel("Alpha")
plt.ylabel("R^2 score")
plt.draw()
        
print("Best score: {:.6f}".format(best_score))
print("Best parameters: {}".format(best_parameters))
```

```python
# Lasso (L1)
best_score = 0
lasso_train_scores = []
lasso_test_scores = []

for alpha in range_01:
    lasso = Lasso(alpha=alpha).fit(X_train, y_train)
    score = lasso.score(X_test, y_test)
    lasso_test_scores.append(score)
    lasso_train_scores.append(ridge.score(X_train, y_train))
    if score > best_score:
        best_score = score
        best_parameters = {'alpha': alpha}


fig1=plt.figure(figsize=(14,3))
fig1.show()
ax1=fig1.add_subplot(121)
ax1.plot(range_01,lasso_test_scores,c='b',ls='-',label='Test')
plt.legend(loc=2)
plt.title("Lasso test regularization")
plt.xlabel("Alpha")
plt.ylabel("R^2 score")
ax2=fig1.add_subplot(122)
ax2.plot(range_01,lasso_train_scores,c='r',ls='-',label='Training')
plt.legend(loc=2)
plt.title("Lasso training regularization")
plt.xlabel("Alpha")
plt.ylabel("R^2 score")
plt.draw()

fig2=plt.figure(figsize=(14,3))
fig2.show()
ax=fig2.add_subplot(111)
ax.plot(range_01,lasso_test_scores,c='b',ls='-',label='Test')
ax.plot(range_01,lasso_train_scores,c='r',ls='-',label='Training')
plt.legend(loc=2)
plt.title("Lasso regularization")
plt.xlabel("Alpha")
plt.ylabel("R^2 score")
plt.draw()

        
print("Best score: {:.6f}".format(best_score))
print("Best parameters: {}".format(best_parameters))
```

```python
# Ridge Regression vs Lasso comparison

fig=plt.figure(figsize=(14,3))
fig.show()
ax=fig.add_subplot(111)
ax.plot(range_01,ridge_test_scores,c='b',ls='-',label='Ridge (L2) Test')
ax.plot(range_01,ridge_train_scores,c='b',ls='--',label='Ridge (L2) Training')
ax.plot(range_01,lasso_test_scores,c='r',ls='-',label='Lasso (L1) Test')
ax.plot(range_01,lasso_train_scores,c='r',ls='--',label='Lasso (L1) Training')
plt.legend(loc=2)
plt.title("Regularization comparison")
plt.xlabel("Alpha")
plt.ylabel("R^2 score")
plt.draw()
```

```python
# Elastic Net (L1/L2 tradeoff)
best_score = 0
en_scores = []

for alpha in range_01:
    row = []
    for l1_ratio in range_01:
        en = ElasticNet(alpha=alpha, l1_ratio=l1_ratio).fit(X_train, y_train)
        score = en.score(X_test, y_test)
        row.append(score)
        if score > best_score:
            best_score = score
            best_parameters = {'alpha': alpha, 'l1_ratio': l1_ratio}
    en_scores.append(row)
        

scores = np.array(en_scores).reshape(len(en_scores), len(en_scores))
# plot the mean cross-validation scores
mglearn.tools.heatmap(scores, xlabel='Alpha', xticklabels=range_01,
ylabel='L1 ratio', yticklabels=range_01, cmap="viridis");

print("Best score: {:.6f}".format(best_score))
print("Best parameters: {}".format(best_parameters))
```

<strong>Overview of the optimized performances:</strong><br>
Linear regression -> Best score: 0.949570 (for degree = 2)<br>
Ridge -> Best score: 0.950847 (for alpha = 1)<br>
Lasso -> Best score: 0.950885 (for alpha = 0.1)<br>
Elastic Net -> Best score: 0.950885 (for alpha = 0.1, l1_ratio = 1)<br>

We tried to improve the Linear Regression performance by adding flexibility, through additional polynomial features. The optimal score, which is significatively better than the default lineal model score, was found for a polynomial degree of 2.

When using Ridge Regression, which adopts L2 regularization, the best R^2 score is obtained for the maximal value of alpha (1), meaning that the model benefits from a very strong regularization to reduce the variance of estimates. Contrarily, in the Lasso model (based on L1 resulatization) the optimal value is found for a very low but not minimale value of alpha (0.1), implying that a stronger regularization is counterproductive.

In the Elastic Net model, the optimal value corresponds to a L1 ratio of 1 (full L1 penalty) and an alpha value of 0.1, which brings us back to the same properties and same R^2 score of the Lasso model. As visible from the heatmap, the worst performances are located in the top-left corner, which represents L1 regularization for a high alpha parameter, while all other corners of the map show fairly high R^2 values.

L2 prefers to have many small values, while L1 gives a better performance for many values close to 0 and some very large values, making it robust to ouliers. Interpreting our results, which finds in L1 regularization the best option for our model, we could probably achieve an even better result with a more detailed preprocessing of the data, focused on outliers removal.

2 . A _validation curve_ can help you understand when a model starts under- or overfitting. It plots both training and test set error as you change certain characteristics of your model, e.g. one or more hyperparameters. Build validation curves for gradient boosting, evaluated using AUROC, by varying the number of iterations between 1 and 500. In addition, use at least two values for the learning rate (e.g. 0.1 and 1), and tree depth (e.g. 1 and 4). This will yield at least 4 curves. Interpret the results and provide a clear explanation for the results. When is the model over- or underfitting? Discuss the effect of the different combinations learning rate and tree depth and provide a clear explanation.
- While scikit-learn has a `validation_curve` function, we'll use a modified version (below) that provides a lot more detail and can be used to study more than one hyperparameter. You can use a default train-test split.

```python
# Plots validation curves for every classifier in clfs. 
# Also indicates the optimal result by a vertical line
# Uses 1-AUROC, so lower is better
def validation_curve(clfs, X_test, y_test, X_train, y_train):
    for n,clf in enumerate(clfs):
        test_score = np.empty(len(clf.estimators_))
        train_score = np.empty(len(clf.estimators_))

        for i, pred in enumerate(clf.staged_decision_function(X_test)):
            test_score[i] = 1-roc_auc_score(y_test, pred)

        for i, pred in enumerate(clf.staged_decision_function(X_train)):
            train_score[i] = 1-roc_auc_score(y_train, pred)

        best_iter = np.argmin(test_score)
        learn = clf.get_params()['learning_rate']
        depth = clf.get_params()['max_depth']
        test_line = plt.plot(test_score,
                             label='learn=%.1f depth=%i (%.2f)'%(learn,depth,
                                                                 test_score[best_iter]))

        colour = test_line[-1].get_color()
        plt.plot(train_score, '--', color=colour)
        
        plt.xlabel("Number of boosting iterations")
        plt.ylabel("1 - area under ROC")
        plt.axvline(x=best_iter, color=colour)
        
    plt.legend(loc='best')
```
